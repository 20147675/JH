```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import os 
import warnings 

warnings.filterwarnings('ignore')



```

```python
data= pd.read_csv('/kaggle/input/employee-data/HR-Employee-Attrition.csv') 
data
```

```python
sns.regplot(x=data['JobLevel'], y=data['TotalWorkingYears'])
```

```python
sns.lmplot(x="EmployeeNumber", y="DistanceFromHome", hue="Attrition", data=data)
fig= plt.gcf()
fig.set_size_inches(10,10)
```

```python
sns.lmplot(
    data=data, x="EmployeeNumber", y="DistanceFromHome", col="Attrition", hue="Attrition",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```
